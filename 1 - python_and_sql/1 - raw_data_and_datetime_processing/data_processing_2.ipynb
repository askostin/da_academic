### Загрузка библиотек

```python
import pandas as pd
```

### Загрузка таблицы

```python
df = pd.read_excel('./tables.xlsx')
df = df.dropna(axis = 0, how = 'all').reset_index(drop = True)
```

```python
df
```

```python
# List of indexes for rows, which contains column names.
# For very first table column names are already in the header of the table,
#   they are not in the 0-th row -- for this table we need to find the last row
#   just before the header of the next table.
listof_indexes = [0]

# row[[isinstance(elt, str) for elt in row]]

for index, row in df.iterrows() :
    if (sum([isinstance(elt, str) for elt in row]) > 0) :
        listof_indexes.append(index)
        
# print(listof_indexes)
# print(len(listof_indexes))

# df1 = df.iloc[]
number_of_tables = len(listof_indexes)
listof_indexes.append(df.shape[0]) # Add the index of the last row in the table to the list

tables = []
sum = 0

# i is the number of the sub-table
for i in range(number_of_tables) :
    start = listof_indexes[i] #index of row to begin new sub-table
    end = listof_indexes[i + 1] #index of row which is the last in the new sub-table
    df1 = df.iloc[start:end,].dropna(axis = 1, how = 'all').reset_index(drop = True).fillna(0)
    if i > 0 :
        df1.columns = df1.iloc[0,:].to_list()
        df1 = df1.loc[1:,]
    tables.append(df1)
#     print(df1)
#     print('******************')

result = pd.concat(
    tables,
    axis = 0,
    join = 'outer',
    ignore_index = True
)
```
